Numpy Provides us Features/Functions to Create and Manipulate

---

Pandas is Built on Top of Numpy. it uses ND Array Internally to
store Data.

Features of ND Array are available in Pandas , not Whole Numpy.(math part)

Pandas has Builtin Analytical Functions that help us Analyse Data and
Extract Facts/patterns from Data.

Summary: Pandas provides functionatlies primarily aimed at Data
Analysis.

# Let us learn how to Handle Multiple Conditions in Pandas

---


```python
nba.head()
```

```python
# Objective: you are asked to Find Players from 'Texas' College Playing from 'PG' Position.
```

```python
# Regular Approach

c1 = nba.College == 'Texas'

c2 = nba.Position == 'PG'

nba[ c1 & c2 ]     # & is Logical AND
```

```python
# Direct Expression

nba[ (nba.College == 'Texas') & (nba.Position == 'PG')]

# please note: enclosing conditions in bracket is Mandatory
```

```python
# Using OR Operator

# objective: you are asked to find players playing from either 'SF' or 'C' Position

c1 = nba.Position == 'SF'
c2 = nba.Position == 'C'

nba[ c1 | c2 ].head(10)     # here | is Logical OR
```

```python
b1= nba.Position=="SF"
b2=nba.Position =="C"
nba[b1|b2]
```

```python
# using Not Operator

# you can freely use != sign for Not Behavior
```

```python
# fetching Rows where Salary Contains NaN
```

```python
nba[ nba.Salary.isna() ]
```

```python
# Negation i.e ~ Opeartor
# this reverses the given condition
```

```python
nba[  ~ nba.Salary.isna() ] # fetch players whose salary is not NaN
```

```python
# # Find Players Whose Age is between 22 to 24
```

```python
nba[  nba.Age.between(22,24) ]
```
